## Exploratory Data Analysis (EDA) and Loading Cleaned Data into PostgreSQL for Further Analysis

```python
# import dependencies
import pandas as pd
import numpy as np

# load CSV
file_path = r"C:\Users\anant\Desktop\data analytics\Walmart-Sales-End-to-End-Data-Analytics\Walmart.csv"
df = pd.read_csv(file_path)

# quick check
df.head()
```

```python
# Shows dataset structure, data types
df.info()
```

```python
# Counts total duplicate rows in the dataset
df.duplicated().sum()
```

```python
# Removes duplicate rows from the dataset
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

```python
# Shows count of missing values in each column
df.isna().sum()
```

```python
# Removes rows containing any missing (NaN) values
df.dropna(inplace=True)
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
# Removes '$' symbol, converts unit_price to float, and previews the data
df['unit_price']=df['unit_price'].str.replace('$','').astype(float)
df.head()
```

```python
df.dtypes
```

```python
df['total']=df['unit_price']*df['quantity']
df.head()
```

```python
# import dependency
from sqlalchemy import create_engine
```

```python
# Create database connection
engine = create_engine(
"postgresql+psycopg2://postgres:Bisag%40123@localhost:5432/walmart"
)
```

```python
# load dataframe to SQL table
df.to_sql(
name='walmart_sales',
con=engine,
if_exists='replace', # use 'append' if table already exists
index=False
)


print("Data loaded successfully into walmart database")
```
